```python
#python program to display list of respected former presidents of India(i.e. Name , Term of office) 

from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
president_name=[]
for i in soup.find_all('div',class_='presidentListing'):
    president_name.append(i.find("h3").text)
president_name    
```

```python
president_office_term=[]
for i in soup.find_all('div',class_='presidentListing'):
    president_office_term.append(i.find("p").text.split()[4:10])
president_office_term
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({'President_Name':president_name,'Office_Term':president_office_term})
data
```
